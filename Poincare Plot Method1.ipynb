```python
"""
Lazım olan paketlerimizi dahil ediyoruz
"""
%matplotlib inline
import numpy as np
import random 
import matplotlib.pyplot as plt
```

```python
"""
Örnekler için zaman serisi verileri oluşturuyoruz
"""
x = np.arange(360)
x = np.array(x,float)*np.pi/180
ranSer = np.random.rand(360) # Rastgele değerlerle üretilmiş seri
sinSer = np.sin(x) # Sinüs fonksiyonu ile üretilmiş seri 
cosSer = np.cos(x) # Cosinüs fonksiyonu ile üretilmiş seri
multiSer = np.sin(x)*np.sin(x)-np.cos(3*x)+5  # sin(x)^2 - cos(3x) + 5 fonksiyonu ile üretilmiş seri  
multiSer2 = np.sin(x*x)-2*np.cos(x) + 3 # sin(x^2) - 2* cos(x) + 3 fonksiyonu ile üretilmiş seri  
```

```python
"""
Ekranda gösterebilmek için fonksiyonlar tanımlayalım
"""
def showTimeSeries(x,ts):
    plt.plot(ts,x)
    return

def showHist(ts,bin = 20):
    plt.hist(ts,bin)
    return

def timeS2Poincare(ts):
    x = ts[0:-1]
    y = ts[1:]
    return x,y

def showPoincare(ts):
    x,y = timeS2Poincare(ts)
    plt.plot(x,y,'r.')
    return
```

```python
showTimeSeries(ranSer,x)
```

```python
showTimeSeries(sinSer,x)
```

```python
showTimeSeries(cosSer,x)
```

```python
showTimeSeries(multiSer,x)
```

```python
showTimeSeries(multiSer2,x)
```

```python
showPoincare(sinSer)
```

```python
showPoincare(multiSer)
```

```python
showPoincare(multiSer2)
```
